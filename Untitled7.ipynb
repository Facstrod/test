```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier 
from sklearn.neural_network import MLPRegressor 


data = pd.DataFrame(pd.read_excel('TOO_оптим.xls', sheet_name='Лист2'))
data.index +=1

data = data.loc[:,~data.columns.str.contains('^Unnamed')]
```

```python
X = data.iloc[:, 3:7]
y = data.iloc[:,7:8]
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
mlp = MLPRegressor(hidden_layer_sizes=[5, 5, 10], max_iter=1000)
mlp.fit(X_train, y_train.values.ravel())
predictions = mlp.predict(X_test)
df_final = y_test
df_final['Predicted'] = predictions
```

```python
import matplotlib.pyplot as plt

plt.plot(np.array(df_final['Predicted']), label = 'predicted')
plt.plot(np.array(y_test['W']), label = 'real')
plt.legend()
```

```python
df_final['error'] = (df_final['W']-df_final['Predicted'])/df_final['W']
df_final
```
